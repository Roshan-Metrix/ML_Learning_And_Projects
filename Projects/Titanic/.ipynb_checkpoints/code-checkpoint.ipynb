# Titanic Survival Prediction 

### Data Cleaning, Preprocessing and Filling

```python
import pandas as pd

df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')
df1.head()
```

```python
df2.head()
```

```python
df1.shape
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df1.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
df2.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df1['familySize'] = df1['SibSp'] + df1['Parch'] + 1
df2['familySize'] = df2['SibSp'] + df2['Parch'] + 1
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.drop(['SibSp','Parch'],axis=1,inplace=True)
df2.drop(['SibSp','Parch'],axis=1,inplace=True)
```

```python
df1.head()
```

```python
df1['Sex'] = df1['Sex'].replace({'male': 0, 'female': 1})
df2['Sex'] = df2['Sex'].replace({'male': 0, 'female': 1})
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['Embarked'] = df2['Embarked'].fillna(df2['Embarked'].mode()[0])
df2['Embarked'] = df2['Embarked'].fillna(df2['Embarked'].mode()[0])
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df2['Fare'] = df2['Fare'].fillna(df2['Fare'].median())
```

```python
df2.head()
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].median())
df2['Age'] = df2['Age'].fillna(df2['Age'].median())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Embarked'] = le.fit_transform(df1['Embarked'])
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Embarked'] = le.fit_transform(df1['Embarked'])
df2['Embarked'] = le.fit_transform(df2['Embarked'])
```

```python
df1.head()
```

```python
df2.head()
```

```python
X = df1.drop('Survived',axis=1)
y = df1['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Model Training

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

### Prediction

```python
predictions = model.predict(df2)

output = pd.DataFrame({'PassengerId': df2.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
