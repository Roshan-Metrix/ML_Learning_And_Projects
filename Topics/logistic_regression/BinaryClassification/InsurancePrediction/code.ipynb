```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
df
```

```python
plt.scatter(df["age"],df["have_insurace"],marker="+",color="red")
plt.xlabel("age",fontweight="bold")
plt.ylabel("Have Insurance",fontweight="bold")
plt.title("Age v/s Having Insurance",fontweight="bold")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(df[['age']],df['have_insurace'], test_size = 0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_test
```

```python
model.predict(x_test)
```

```python
model.predict([[3]])
```

```python
model.predict_proba(x_test)
```

```python
model.score(x_test, y_test)
```
