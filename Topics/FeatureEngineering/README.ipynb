# Feature Engineering
Feature Engineering in Machine Learning means turning raw data into better information that a machine can understand more easily.

## Think of it like preparing ingredients before cooking
```
   Good ingredients → better food
   Good features → better ML predictions
```

## Simple Example
Suppose you want to predict house prices.

**Raw data:**

```
Area = 2000 sq ft
Bedrooms = 3
Built year = 2010
```

**Feature Engineering creates smarter features like:**

```
House Age = Current Year − Built Year
Price per Bedroom
Area per Room
```


### Conclusion:
##### "Feature Engineering is the art of helping machines see important patterns in data."
