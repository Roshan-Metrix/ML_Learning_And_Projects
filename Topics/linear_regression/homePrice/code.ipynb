```python
from sklearn import linear_model
```

```python
new_df= df.drop('price',axis='columns')
new_df
```

```python
model = linear_model.LinearRegression() #creating object for LinearRegression
model.fit(new_df,df.price)
```

```python
model.predict([[3000]]) # predicting the price of area 3000
```

```python
model.coef_
```
