```python
# PCA - is a process of figuring out most important features or principal components that has the mose impact on the target variable.
# PCA is called dimensionality reduction technique as it can help us reduce dimentions
```

```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
dataset = load_digits()
dataset.keys()
```

```python
dataset.data[0] # give 1-D array data of data set 
```

```python
dataset.data[0].reshape(8,8) # it make the 1-D into 2-D
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.gray()
plt.matshow(dataset.data[0].reshape(8,8))
plt.matshow(dataset.data[1].reshape(8,8))
```

```python
dataset.target[1]
```

```python
dataset.feature_names
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names) # changing column name as feature name
df.head()
```

```python
df.describe()
```

```python
X = df
y = dataset.target
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
X.head(3)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
# the above PCA makes 64 columns to 29 columns which makes fast
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
# Hence proved , decresing the no. of columns from 64 to 29 gives same 0.97 (97%)
```
