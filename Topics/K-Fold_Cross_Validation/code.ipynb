```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size=0.3)
```

```python
# Below is example of checking model accuracy for digits individually 
```

```python
# By Logistic Regression
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test) 
```

```python
# By SVC
svm = SVC()
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
# By Random Forest
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Now by K-Fold classification
```

```python
from sklearn.model_selection import KFold
Kf = KFold(n_splits=3)
Kf
```

```python
for train_index, test_index in Kf.split([1,2,3,4,5,6,7,8,9]):
    print(train_index,test_index)
```

```python
def get_score(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
get_score(LogisticRegression(),X_train,X_test,y_train,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)

scores_logistic = []
scores_svm = []
scores_rf = []

for train_index, test_index in folds.split(digits.data, digits.target):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index],digits.target[train_index], digits.target[test_index]

    scores_logistic.append(get_score(LogisticRegression(),X_train,X_test,y_train,y_test))
    scores_svm.append(get_score(SVC(),X_train,X_test,y_train,y_test))
    scores_rf.append(get_score(RandomForestClassifier(),X_train,X_test,y_train,y_test))
```

```python
scores_logistic
```

```python
scores_svm
```

```python
scores_rf
```

```python
# doing all k-ford work in a few line 
from sklearn.model_selection import cross_val_score
```

```python
# For LogisticRegression
cross_val_score(LogisticRegression(),digits.data,digits.target)
```

```python
# For SVM
cross_val_score(SVC(),digits.data,digits.target)
```

```python
# For RandomForestClassifier
cross_val_score(RandomForestClassifier(),digits.data,digits.target)
```

```python
# with cross_val_score we can find the actual result of the models 
```

```python
# Putting more parameters for finding actual res for RandomForestClassifier
Rscores = cross_val_score(RandomForestClassifier(n_estimators=10),digits.data,digits.target,cv=10)
Rscores

# n_estimator means no. of Trees (True,False tree)
# cv means no. of value required to return
```

```python
np.average(Rscores)
```
